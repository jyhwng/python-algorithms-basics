5.10 Shell sort
===
삽입 정렬의 단점을 보완하기 위해 고안한 방법  
삽입 정렬의 경우 정렬할 항목이 최종적으로 정렬되어 저장되어지는 위치에서 거리가 멀리 떨어져 있는 만큼 이웃 항목들과 더 많은 비교를 수행해야 함으로 정렬의 효과가 떨어진다.   
따라서 **삽입 정렬은 정렬해야 하는 항목이 어느 정도 정렬이 되어 있는 상태에서 가장 좋은 효과**를 볼 수 있는데, 쉘 정렬은 이러한 장점을 이용하고자 전체 N개라는 하나의 배열을 일정한 간격으로 나누어 보다 작은 자료의 개수로 이루어진 하위 배열을 각각 삽입 정렬 한 뒤 다시 더 좁은 간격으로 하위 배열로 나누고 최종적으로 전체 배열을 삽입 정렬하여 정렬하게 된다.

>1. N개의 항목으로 이뤄진 하나의 배열을 일정한 간격을 둬 나눈 여러 개의 하위 배열로 구성한다.
>2. 여러 개의 하위 배열을 각각 삽입 정렬한다.
>3. 보다 작은 간격을 두어 여러 개의 하위 배열을 구성한다.

Step1. N개의 항목으로 이뤄진 하나의 배열을 일정한 간격을 둬 나눈 여러 개의 하위 배열로 구성한다. (interval = 3)

![shell sort1](http://interactivepython.org/runestone/static/pythonds/_images/shellsortA.png)
![shell sort2](http://interactivepython.org/runestone/static/pythonds/_images/shellsortB.png)

interval을 3으로 했을 때, 총 3개의 하위 배열이 생기고, 이것들은 Sub Directory 내부에서 삽입 정렬을 수행합니다.

![shell sort1](http://interactivepython.org/runestone/static/pythonds/_images/shellsortA.png)
![shell sort2](http://interactivepython.org/runestone/static/pythonds/_images/shellsortB.png)

예제 소스는 위와 조금 다른 interval로 shell sort를 수행합니다.  
위의 interval은 3으로 시작하지만 아래 예제는 $n/2$으로 시작하여 그 다음은 $n/4$ 그리고 interval이 1이 될 때 까지 반복합니다.

```python
def shellSort(alist):
    sublistcount = len(alist)//2  
    while sublistcount > 0:

      for startposition in range(sublistcount):
        gapInsertionSort(alist,startposition,sublistcount)

      print("After increments of size",sublistcount,
                                   "The list is",alist)

      sublistcount = sublistcount // 2

def gapInsertionSort(alist,start,gap):
    for i in range(start+gap,len(alist),gap):

        currentvalue = alist[i]
        position = i

        while position>=gap and alist[position-gap]>currentvalue:
            alist[position]=alist[position-gap]
            position = position-gap

        alist[position]=currentvalue

alist = [54,26,93,17,77,31,44,55,20]
shellSort(alist)
print(alist)
```

5.11 Merge Sort
===
병합정렬은 정렬할 리스트를 반씩 쪼개나가며 좌측과 우측 리스트를 계속하여 분할해 나간 후 각 리스트내에서 정렬 후 병합(merge) 정렬 후 병합하는 과정을 통해 정렬하는 알고리즘이다. 

>performance: $O(nlogn)$  
>space complexity: $O(n)$

병합 정렬의 핵심은 반복적으로 목록을 분할하는 **재귀 알고리즘** 이라는 점이다.

먼저 병합 정렬의 순서는 아래 2가지 이다.
1. 리스트의 원소가 1개가 될 때 까지 n개의 sublist로 계속해서 나눈다 .
2. 분리된 리스트 두개를 비교하여 정렬한다.

![merge sort1](http://interactivepython.org/runestone/static/pythonds/_images/mergesortA.png)

![merge sort2](http://interactivepython.org/runestone/static/pythonds/_images/mergesortB.png)


병합정렬은 가장 많이 쓰이는 정렬 알고리즘 중 하나 이며 시간복잡도는 최소 $O(nlogn)$ 을 보장한다.    

![merge sort3](http://ejklike.github.io/assets/20170301/mergesort.gif)

두 부분으로 분해하는데 $O(logn)$(이진탐색 참고)이고, 데이터 병합이 $O(n)$이므로, 정렬 상태와 무관하게 언제나 $O(nlogn)$이다. 데이터 크기만한 메모리가 더 필요한 게 단점이다.

```python
def mergeSort(alist):
    print("Splitting ",alist)
    if len(alist)>1:
        mid = len(alist)//2
        lefthalf = alist[:mid] # 분할 이후 왼쪽 리스트
        righthalf = alist[mid:] # 분할 이후 오른쪽 리스트

        mergeSort(lefthalf)
        mergeSort(righthalf)

        i=0 lefthalf의 Index Num을 저장할 변수
        j=0 # righthalf의 Index Num을 저장할 변수
        k=0 # alist의 Index Num을 저장할 변수
        while i < len(lefthalf) and j < len(righthalf):
            if lefthalf[i] < righthalf[j]:
                alist[k]=lefthalf[i]
                i=i+1
            else:
                alist[k]=righthalf[j]
                j=j+1
            k=k+1

        while i < len(lefthalf):
            alist[k]=lefthalf[i]
            i=i+1
            k=k+1

        while j < len(righthalf):
            alist[k]=righthalf[j]
            j=j+1
            k=k+1
    print("Merging ",alist)

alist = [54,26,93,17,77,31,44,55,20]
mergeSort(alist)
print(alist)
```

5.12 Quick Sort
===
퀵정렬은 real-world 데이터에서 빠르다고 알려져 있어 가장 많이 쓰는 정렬알고리즘이다.
퀵정렬은 pivot을 선정하여 pivot을 기준으로 좌측과 우측으로 pivot보다 작은값은 왼쪽 pivot보다 큰값은 오른쪽으로 재배치를 하고 계속하여 분할하여 정렬하는 알고리즘이다.
최악의 경우에는 $O(n²)$의 비교를 수행하지만 일반적으로 $O(nlogn)$의 시간복잡도를 가진다.

![Quick Sort](http://interactivepython.org/runestone/static/pythonds/_images/firstsplit.png)
![Quick Sort2](http://interactivepython.org/runestone/static/pythonds/_images/partitionA.png)
![Quick Sort3](http://interactivepython.org/runestone/static/pythonds/_images/partitionB.png)

![Quick Sort4](http://ejklike.github.io/assets/20170301/quicksort.gif)

>1. 배열, 시작위치, 끝 위치를 필요로 한다.
>1. 시작위치가 끝위치와 같거나 크면(클 일은 사실 없음) 이미 정렬된 상태로 간주한다.
>1. 첫 위치가 피봇이 된다. 
>1. 첫 위치의 다음 인덱스가 leftmark, 끝 위치는 rightmark 가 된다. 
>1. leftmark는 피봇값보다 큰 값을 발견할 때까지 증가
>1. right는 피봇값보다 작은 값을 발견할 때까지 감소 
>1. 5, 6을 수행한 후 두 마크가 만나지 않았다면, 두 마크 간의 값을 교환한다. 그리고 5, 6을 계속 진행한다. 
>1. 만약 교차가 일어났다면 즉 right 마크의 인덱스값이 left 마크보다 작다면? right마크와 피봇의 값을 교환한다. 
>1. 이제 0 ..> 피봇, 피봇 <.. 끝 까지의 범위에 대해 각각 퀵 소트를 시전한다. 이 재귀 알고리듬은 2를 베이스로하면서 종료하게 된다. 

```python
def quickSort(alist):
   quickSortHelper(alist,0,len(alist)-1)

def quickSortHelper(alist,first,last):
   if first<last:

       splitpoint = partition(alist,first,last)

       quickSortHelper(alist,first,splitpoint-1) #재귀적으로 퀵 정렬 알고리즘 재 호출 (왼쪽)
       quickSortHelper(alist,splitpoint+1,last) #재귀적으로 퀵 정렬 알고리즘 재 호출 (오른쪽)


def partition(alist,first,last):
   pivotvalue = alist[first] #피벗값을 왼쪽 첫번째 값으로 지정

   leftmark = first+1 #피벗 다음 값부터 leftmark를 수행
   rightmark = last #rightmark는 인덱스의 마지막

   done = False # 수행 여부를 저장하는 변수
   while not done:

       while leftmark <= rightmark and alist[leftmark] <= pivotvalue: # 
           leftmark = leftmark + 1

       while alist[rightmark] >= pivotvalue and rightmark >= leftmark:
           rightmark = rightmark -1

       if rightmark < leftmark:
           done = True
       else:
           temp = alist[leftmark]
           alist[leftmark] = alist[rightmark]
           alist[rightmark] = temp

            #치환
   temp = alist[first] 
   alist[first] = alist[rightmark]
   alist[rightmark] = temp


   return rightmark

alist = [54,26,93,17,77,31,44,55,20]
quickSort(alist)
print(alist)
```

데이터를 절반씩 쪼갠다면 $O(logn)$이고 $n$개의 값을 피봇과 비교하므로 $O(nlogn)$이다. 피봇 선정 기준에 따라 쪼개는 위치가 한 쪽에 치우쳐 있을 수도 있다 (예: 데이터가 이미 정렬되어 있고, 맨 끝 원소를 피봇으로 선정하는 경우). 이 때는 쪼개는 비용이 $O(n)$이므로 퀵정렬의 시간복잡도가 $O(n2)$까지 증가한다. 재귀 호출을 위한 스택을 사용하므로 추가 메모리가 필요하다.
